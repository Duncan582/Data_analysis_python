## 1.Series(一维数据:时间序列/传感器读数)(类似于Excel中的单列)(要求同质化数据类型)


### I.创建方法

```python
import pandas as pd
import numpy as np
```

```python
# 创建Series
s = pd.Series([1, 2, 3, 4, 5], index = ['A', 'B', 'C', 'D', 'E'])
# This includes the indexs, the values and the dtype!!!
print(s)
# 自定义索引
s = pd.Series([1, 2, 3, 4, 5], index = [1, 2, 3, 4, 5])
print(s)
# 定义name
s = pd.Series([1, 2 ,3, 4, 5], index = [1, 2, 3, 4, 5], name = '月份')
# name can be set and showed!!!
print(s)
```

```python
# 通过字典创建Series
s = pd.Series({'a':10, 'b':20, 'c':30, 'd':40, 'e':50, 'f':60 })
print(s)
s1 = pd.Series(s, index = ['a', 'c'])
# 取已有Series的特定索引与值
print(s1)
```

### II.Series的属性
index:索引

values:值

loc[]:显式索引

iloc[]:隐式索引(位置)

at[]:使用标签访问单个元素

iat[]:使用位置访问单个元素

dtype/dtypes:数据类型

```python
print(s)
print(s.index)
print(s.values)
print(s.shape, s.ndim, s.size)
print(s.dtype)
print(s.name) # None means not defined
print(s.loc['a':'c']) # 通过索引找到对应的值; 这里包含右端的end
print(s.iloc[0:2]) # 通过隐式索引(真正的索引数)找到对应的值; 注意不包含右端end
print(s.at['a']) # 通过索引找到对应的值
print(s.iat[0]) # 通过隐式索引(真正的索引数)找到对应的值
```

```python
print(s)
# 访问数据
print(s['a']) # 如果这里填入数字(position),会出现Warning:实际上不可以通过s[position]取值,只能用s[label]
# bool索引
print(s[s < 30])
print(s.head()) # 返回前五行数据(行数可调)
print(s.tail(2)) # 返回后五行数据(行数可调)
```

### III.常用方法
isin():判断元素是否在参数集合中

isna():判断是否为缺失值(NaN或None)

mode():众数,可返回多个

quantile(q):分位数,q取0-1

describe():直接给出常见统计信息:count,mean,std,min,25%,50%,75%,max

value_counts():每个唯一值出现次数

count():非缺失值数量

nunique():唯一值个数(去重)

unique():获得去重后的值数组

drop_duplicates():去除重复项

sample():随机抽样

sort_index():按索引排序

sort_values():按值排序

replace():替换值

keys():返回Series的索引对象

```python
# 常见函数
s = pd.Series([10, 2, np.nan, None, 3, 4, 5], index = ['A', 'B', 'C', 'D', 'E', 'F', 'G'], name = 'data')
print(s)
```

```python
s.head(3)
s.tail(6)
```

```python
# 获取元素个数
print(s.count())
# 查看所有描述性信息
s.describe()
```

```python
print(s.keys()) # 方法
print(s.index) # 属性
```

```python
# 检查每个元素是否为缺失值
print(s.isna())
s.isna()
```

```python
s.isin([4, 5]) # 检查每个元素是否在给定参数集合中
```

```python
print(s.sort_values())
print(s.mean())
print(s.std())
print(s.var())
print(s.min())
print(s.quantile(0.25))
print(s.median())
print(s.quantile(0.75))
print(s.quantile(0.8))
# s.quantile(0.8) = (0.8 * 4 - 3) * (10 - 5) + 5
print(s.max())
```

```python
# 众数
s['H'] = 4
print(s.sort_index())
print(s.mode()) # 会输出所有的众数
```

```python
# 各个元素计数
print(s.value_counts())
```

```python
# 去重
print(s.drop_duplicates()) # return a Series
print(s.unique()) # return an array
print(s.nunique()) # 获得去重后元素的个数
```

```python
# 排序:值/索引
s.sort_index()
s.sort_values()
```
