# hw1.学生成绩统计
 创建一个包含10名学生成绩的Series,成绩在50-100,计算平均分最高分最低分,并找出高于平均分的人数

```python
import numpy as np
import pandas as pd
```

```python
np.random.seed(42)
scores = pd.Series(np.random.randint(10, 101, 10), index = ['学生' + str(i) for i in range (1, 11)])
print(scores)
print('平均分:', scores.mean())
print('最高分:', scores.max())
print('最低分:', scores.min())
# print('高于平均分人数:', len(scores[scores > scores.mean()]))
print('高于平均分人数:', scores[scores > scores.mean()].count())
```

# hw2:温度数据分析
找出超过30℃的天数,计算平均温度,将温度从高到低排序,找出温度变化最大的两天

```python
temps = pd.Series([28, 31, 29, 32, 30, 27, 33], 
                  index = ['周一', '周二', '周三', '周四', '周五', '周六', '周日'])
print('温度超过30度的天数:', len(temps[temps>30]))
print('平均温度:', temps.mean())
print('从高到低排序结果:', '\n', temps.sort_values(ascending = False))
temp1 = temps.diff().abs() # 计算Series数组值的变化量
print(temp1.sort_values(ascending = False)) # 降序排序
print('温度变化最大的两天:', '\n',
      temp1.sort_values(ascending = False).keys()[:2].tolist())
```

# hw3:股票价格分析
计算每日收益率(当日收盘价 / 前日收盘价 - 1), 找出收益率最高和最低的日期, 计算波动率(收益率的标准差)

```python
prices = pd.Series([102.3, 103.5, 105.1, 104.8, 106.2, 107.0, 106.5, 108.1, 109.3, 110.2], 
                   index = pd.date_range('2023-01-01', periods = 10, freq = 'D')) # 日期序列生成
print(prices)
```

```python
# 计算每日收益率
print('收益率:', prices.pct_change(), sep = '\n') # percent = prices[i] / prices[i - 1] - 1
```

```python
percent = prices.pct_change()
print('收益率最高日期:', percent.idxmax())
print('收益率最低日期:', percent.idxmin())
print('波动率:', percent.std())
```

# hw3:销售数据分析
给出12个月销售量, 计算季度平均销量, 找出销量最高月份, 计算月环比增长率(每月与上月之比), 找出连续增长超过两个月的月份

```python
# Series日期序列生成
sales = pd.Series([120, 135, 145, 160, 155, 170, 180, 175, 190, 200, 210, 220],
                  index = pd.date_range('2022-01-01', periods = 12, freq = 'MS'))
print(sales)
```

```python
# 季度平均销量
print('季度平均销量', sales.resample('QS').mean(), sep = '\n') # 重新采样
# 销量最高月份
print('销量最高月份:', sales.idxmax())
# 月环比增长率
print('月环比增长率:', sales.pct_change(), sep = '\n')
# 连续增长超过两个月的月份
# 通过增长率连续为正判断
change = sales.pct_change()
ascend = change > 0
print('连续增长超过两个月的月份', ascend[ascend.rolling(3).sum() == 3].keys().tolist(), sep = '\n')
# 滑动窗口求三个月的增长率和,若为3个True,取索引,转列表
```

# hw4:每小时销售数据分析
按天重采样计算每日的销售总额, 计算每天停业时间(8:00 - 22:00)和非营业时间的销售额比例, 找出销售额最高的三小时

```python
np.random.seed(42)
hourly_sales = pd.Series(np.random.randint(0, 100, 24),
                index = pd.date_range('2025-01-01', periods = 24, freq = 'h'))
print(hourly_sales)
```

```python
total = hourly_sales.resample('D')
print('当日销售额:', total.sum(), sep = '\n')
# on_sale = hourly_sales.between_time('8:00','22:00').sum() # 筛选一段时间内的Series(营业时段)
# off_sale = hourly_sales.drop(on_sale.index) # 去除营业时段,得到非营业时段

# Another way
mask = (hourly_sales.index.hour>=8) & (hourly_sales.index.hour<=22)
on_sale = hourly_sales[mask]
off_sale = hourly_sales[~mask]
print(off_sale)
print('营业与非营业时间销售额之比:', on_sale.sum()/(total.sum() - on_sale.sum()), sep = '\n')
```

```python
# 销售额最高的三个小时
print(hourly_sales.nlargest(3).keys().tolist())
```
