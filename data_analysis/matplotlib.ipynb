# 数据可视化
折线图(plot),柱状图,散点图(scatter),条形图(bar),饼图(pie),箱型图(boxplot)

准确高效有效设计美观

matplotlib:基础,灵活,代码多,风格基础

seaborn:基于matplotlib的高级接口,美观,对简单图繁琐

pandas plot:快速图表,调用.plot(),快捷,适合EDA，图表样式较少

```python
import matplotlib.pyplot as plt
from matplotlib import rcParams # 汉字字体支持
rcParams['font.family'] = 'SimHei'
rcParams['font.sans-serif'] = 'SimHei'
```

```python
# 折线图
# 创建图表
plt.figure(figsize = (10, 5))
# 数据
month = ['1月', '2月', '3月', '4月']
sales = [100,150,80,130]
# 绘图
plt.plot(month,sales, 
         label = '产品A',# 图例标签名
         color = 'orange',
         linewidth = 2,
         linestyle = '--',
         marker = 'o') # 数据点标注
# 标题
plt.title('2025年销售趋势', color = 'green', fontsize = 20)
# 坐标轴标签
plt.xlabel('月份', fontsize = 10)
plt.ylabel('销售额(万元)',fontsize = 10)
# 图例
plt.legend(loc = 'upper left') # 添加图例
# 添加网格线
plt.grid(True, alpha = 0.4, linestyle = '--', color = 'blue') 
# x,y轴均有; alpha控制透明度,linestyle='--'代表虚线,linestyle = '-'代表实线
# plt.grid(axis = 'x')
# 设置刻度字体大小
plt.xticks(rotation = 45, fontsize = 15)
plt.yticks(rotation = 0, fontsize = 15)
# 设置y轴范围
plt.ylim(0, 160)
# 数据点显示
for x,y in zip(month, sales):
    plt.text(x,y+2,str(y),ha = 'center', va = 'bottom', fontsize = 10) #ha/va:水平/竖直对齐方式
# 显示图表
plt.show()
```

```python
# 柱状图
# 创建图表
plt.figure(figsize = (10, 5))
# 数据
subjects = ['语文', '数学', '英语', '科学']
scores = [85, 92, 78, 88]
# 绘图(柱状图)
plt.bar(subjects,scores,
        label = 'Ray')
# 添加标题
plt.title('2025成绩',fontsize = 20)
# 添加坐标轴标签
plt.xlabel('科目',fontsize = 10)
plt.ylabel('分数',fontsize = 10)
# 图例
plt.legend(loc = 'upper right')
# 网格线
plt.grid(axis = 'y', alpha = 0.2, color = 'blue', linestyle = '--')
# 刻度字体大小
plt.xticks(rotation = 0, fontsize = 12)
plt.yticks(rotation = 0, fontsize = 12)
# 设置y轴范围
plt.ylim(0,100)
# 在每个数据点显示数值
for x,y in zip(subjects,scores):
    plt.text(x,y+2,str(y),ha = 'center',va = 'bottom',fontsize = 10)
# 显示图表
plt.show()
```

```python
# 条形图
# 创建图表
plt.figure(figsize = (10, 5))
# 数据
countries = ['US', 'CN', 'JP', 'KR', 'Germany']
gdp = [92, 90, 85, 80, 85]
# 绘图(条形图)
plt.barh(countries,gdp,color = 'orange')
# 添加标题
plt.title('2025GDP',fontsize = 20, color = 'red')
# 坐标轴
plt.xlabel('GDP',fontsize = 10)
plt.ylabel('国家',fontsize = 10)
# 自动优化排版
plt.tight_layout()
# 显示图表
plt.show()
```

```python
# 饼图
plt.figure(figsize = (10,5))
# 数据
things = ['learning', 'entertainment', 'sports', 'sleep', 'other']
times = [5, 4, 2, 8, 5]
colors = ['#66b3ff', '#99ff99', '#ffcc99', '#ff9999', '#ff4499']
plt.pie(times, labels = things, 
        autopct = '%.1f%%', # %%可以添加%
        startangle = 90, # 调整初始画图位置
        colors = colors)
plt.title('一天的时间分布', color = 'red', fontsize = 20)
plt.tight_layout()
plt.show()
```

```python
# 环形图
plt.figure(figsize = (10,5))
# 数据
things = ['learning', 'entertainment', 'sports', 'sleep', 'other']
times = [5, 4, 2, 8, 5]
colors = ['#66b3ff', '#99ff99', '#ffcc99', '#ff9999', '#ff4499']
plt.pie(times, labels = things, 
        autopct = '%.1f%%', # %%可以添加%
        startangle = 90, # 调整初始画图位置
        colors = colors, # 配色设置
        wedgeprops = {'width':0.6}, # 圆环宽度设置
        pctdistance = 0.7, # 数字位置
        )
plt.title('一天的时间分布', color = 'red', fontsize = 20)
plt.text(0,-0.06,'总计:\n100%',ha = 'center', va = 'bottom', fontsize = 10)
plt.tight_layout()
plt.show()
```

```python
# 爆炸式饼图
plt.figure(figsize = (10,5))
# 数据
things = ['learning', 'entertainment', 'sports', 'sleep', 'other']
times = [5, 4, 2, 8, 5]
colors = ['#66b3ff', '#99ff99', '#ffcc99', '#ff9999', '#ff4499']
explode = [0.1,0,0,0,0] # 设置突出模块的位置(距离)
plt.pie(times, labels = things, 
        autopct = '%.1f%%', # %%可以添加%
        startangle = 0, # 调整初始画图位置
        colors = colors,
        explode = explode,
        shadow = True, # 加阴影
        )
plt.title('一天的时间分布', color = 'red', fontsize = 20)
plt.tight_layout()
plt.show()
```

```python
# 散点图
plt.figure(figsize = (10,8))
# 数据
import random
x = []
y = []
for i in range(1000):
    temp = random.uniform(0,10)
    x.append(temp)
    temp2 = 2 * temp + random.gauss(0,2)
    y.append(temp2)
# 绘制
plt.scatter(x, y,
            color = 'blue',
            alpha = 0.5,
            s = 20,
            label = '数据'
            )
plt.title('y-x', color = 'red', fontsize = 20)
plt.xlabel('x变量', fontsize = 10)
plt.ylabel('y变量', fontsize = 10)
plt.legend(loc = 'upper left')
plt.grid(True, alpha = 0.1, color = 'blue', linestyle = '--')
plt.xticks(rotation = 0, fontsize = 10)
plt.yticks(rotation = 0, fontsize = 10)
plt.ylim(0, 30)
# 绘制回归线
plt.plot([0,10], [0,20], color = 'orange', linewidth = 2, linestyle = '-')

plt.tight_layout()
plt.show()
```

```python
# 箱型图
data = {
    '语文':[82, 85, 88, 70, 90, 76, 84, 83, 95],
    '数学':[75, 80, 79, 93, 88, 82, 87, 89, 92],
    '英语':[70, 72, 68, 65, 78, 80, 85, 90, 95]
}
plt.figure(figsize = (8,6))
plt.boxplot(data.values(), tick_labels = data.keys())

plt.title("各科成绩分布(箱型图)")
plt.ylabel('分数')
plt.grid(True, axis = 'y', linestyle = '--', alpha = 0.5)
plt.show()
```
