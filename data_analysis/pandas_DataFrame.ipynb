# DataFrame
每行可以用idx寻找, 每一列都有Column名(都是一个Series), 因此每列内部数据类型必须相同

## 1.创建方法

```python
import pandas as pd
import numpy as np
```

```python
# 通过Series创建
s1 = pd.Series([1, 2, 3, 4, 5])
s2 = pd.Series([6, 7, 8, 9, 10])
df = pd.DataFrame({"第一列" : s1, "第二列" : s2})
# 通过字典创建
df = pd.DataFrame(
    {
        "name" : ['makoto', 'futaba', 'yosuke', 'Rei', 'yuuki'],
        "age" : [17, 15, 17, 17, 17],
        "score" : [99.5, 95, 95, 95, 90.5]
    }, index = [1, 2, 3, 4, 5], columns = ["name", "score", "age"]
) # index指定行索引,columns指定列的顺序
print(df)
```

## 2.DataFrame的属性
index:行索引

dtypes:元素类型

columns:列标签

loc[]/iloc[]:显式/隐式索引,按行列标签索引或切片

at[]/iat[]:使用行列位置访问单个元素

T:行列转置

```python
print('行索引:')
print(df.index)
print('列标签:')
print(df.columns)
print('值')
print(df.values)
```

```python
print('维度')
print(df.ndim)
print('形状')
print(df.shape)
print('元素个数')
print(df.size)
print('数据类型')
print(df.dtypes) # return the type of every column
print('转置')
print(df.T)
print(df.T.index) # 转置后index, columns互换
```

```python
# 获取元素
# 行
print(df.loc[1]) # 填入标签(index)
print(df.iloc[0]) # 填入默认的index
# 列
print(df.loc[: , 'name']) # 获取列数据需要行和列的信息
print(df.iloc[: , 0])
```

```python
# 获取单个元素
print(df.at[1 , 'name'])
print(df.iat[0 , 0])
print(df.loc[1 , 'name'])
print(df.iloc[0,0])
```

# 3.访问数据(获取某行/某列的数据)

```python
print(df['name'])
print(df[['name','score']]) # This returns a DataFrame!!! 可用于获取多列数据
print(df.score)
# print(df.loc[:,'name'])
```

```python
# 查看部分数据
print(df.head(2))
print(df.tail(3))
```

```python
# bool索引
print(df[df.score>94])
print(df[(df.score>94) & (df.age > 15)])
```

```python
# 随机抽样
df.sample(3) # 顺序随机
```

# 4.常用方法
duplicated():是否重复, replace():替换值, nlargest:返回某列最大的n条数据, nsmallest()

```python
print(df)
```

```python
print(df.isin(['makoto', 17])) # 查看元素是否包含在参数集合中
print(df.isna())
```

```python
print(df['score'].sum()) # 求某列总和
print(df.score.max())
print(df.score.min())
print(df.score.mean())
print(df.score.median())
print(df.score.mode())
```

```python
print(df.score.std())
print(df.score.var())
print(df.score.quantile(0.25))
```

```python
print(df.describe())
```

```python
print(df.count()) # 返回每一列非缺失值的个数
print(df.value_counts()) # 是否重复
print(df.duplicated(subset = ['age'])) # 比较每条数据和上一条是否重复
print(df.drop_duplicates(subset = ['name']))
```

```python
print(df.replace(17, 20)) # 不改变df 
```

```python
print(df.cumsum())
print(df.cummin(axis = 0)) # 求列的最值, 可设置行(axis = 1)/列,但要求没有字符串才不会报错
```

```python
print(df.sort_index(ascending = False))
print(df.sort_values(by = ['score','age'], ascending = [False, False] ))
```

```python
print(df.nlargest(4 , columns = ['score', 'age']))
print(df.nsmallest(2, columns = ['score', 'age']))
```
