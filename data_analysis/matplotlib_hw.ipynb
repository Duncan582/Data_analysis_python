# hw1:温度分析

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams['font.sans-serif'] = 'SimHei'
```

```python
df = pd.read_csv('weather.csv')

# 绘制气温变化趋势
df['date'] = pd.to_datetime(df['date'])
plt.figure(figsize = (10,10))
plt.plot(df['date'], df['temp_max'], label = '高温变化趋势')
plt.plot(df['date'], df['temp_min'], label = '低温变化趋势')
plt.title('2025某地气温变化趋势', color = 'red', fontsize = 20)
plt.xlabel('日期', fontsize = 12)
plt.ylabel('温度', fontsize = 12)
plt.legend(loc = 'upper left')
```

```python
# 平均气温
df['temp_mean'] = (df['temp_max'] + df['temp_min']) /2
plt.figure(figsize = (10,10))
plt.plot(df['date'], df['temp_mean'], label = '平均气温变化趋势')
plt.title('2025某地气温变化趋势', color = 'red', fontsize = 20)
plt.xlabel('日期', fontsize = 12)
plt.ylabel('温度', fontsize = 12)
plt.legend(loc = 'upper left')
```

```python
# 绘制降水量直方图
plt.hist(df['precipitation'], bins = 5) # 降水量分为5个区间
plt.title('2025年降水量分布', color = 'red', fontsize = 20)
```
