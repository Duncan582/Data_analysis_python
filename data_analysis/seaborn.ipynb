# Seaborn是基于matplotlib的二次封装

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import rcParams
plt.rcParams['font.sans-serif'] = ['SimHei']
penguins = pd.read_csv('penguins.csv')
penguins.dropna(inplace = True)
penguins.info()
```

```python
# 直方图
sns.histplot(data = penguins, x = 'species')
```

```python
# 核密度估计图
# 绘制喙的长度的核密度估计图(描述概率分布的平滑密度曲线图)
sns.kdeplot(data = penguins, x = 'bill_length_mm')
```

```python
sns.histplot(data = penguins, x = 'bill_length_mm', kde = True) 
# 直方图叠加核密度估计图
```

```python
# 计数图
# 显示每个类别在数据集中出现的次数
sns.countplot(data = penguins, x = 'island')
# 直方图用于连续变量，计数图用于分类变量
```

```python
# 散点图
sns.scatterplot(data = penguins, x = 'body_mass_g', y = 'fill_length_mm', hue = 'sex')
# hue:区分颜色的因素
```

```python
# 蜂窝图
sns.jointplot(data = penguins, x = 'body_mass_g', y = 'fill_length_mm', kind = 'hex')
# 设置kind = hex可绘制蜂窝图
```

成对关系图:pairplot; 箱线图:barplot; 条形图:barplot
