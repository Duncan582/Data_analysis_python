## 2.缺失值处理

```python
# 缺失值处理
import numpy as np
s = pd.Series([12, 25, np.nan, None, pd.NA])
df = pd.DataFrame([[1,pd.NA,2], [2,3,5], [None,4,6]], 
                  columns = ['第一列','第二列','第三列'])
# 检查缺失值
print(s.isna())
print(s.isnull().sum())
print(df.isna())
print(df.isnull().sum())
```

```python
# 剔除缺失值
print(s.dropna()) # 删除缺失值位置
print(df.dropna()) # 删除含缺失值的记录
print(df.dropna(how = 'all')) # 仅在所有值均为缺失值时删除记录
print(df.dropna(thresh = 2)) # 至少有n个值不是缺失值就保留记录
print(df.dropna(axis = 1)) # 删除整列的目录(axis取值与其他方法不同)
print(df.dropna(subset = ['第一列'])) # 仅删除指定列中存在缺失值的记录
```

```python
# 填充缺失值
df = pd.read_csv('data_withna.csv')
# 使用字典填入固定的值
print(df.fillna({'favorite food' : 'bread', 'score' : 60})) 
# 使用统计值填充
print(df.fillna({'score' : df['score'].mean()}))
# 用前方的值填充
print(df.ffill())
# 用后方的值填充
print(df.bfill())
```

## 5.数据变形

```python
data = {
    'id' : [1, 2],
    'name' : ['alice', 'bob'],
    'Math' : [90, 85],
    'English' : [88, 92],
    'Science' : [95, 89]
}
df = pd.DataFrame(data)
df.T
df # 宽表(列多于行)
# 宽表转换成长表
df2 = pd.melt(df, id_vars = ['id', 'name'], var_name = '科目', value_name = '分数')
df2.sort_values('name')
# id_vars:不变化的列, var_name:变化的列组成的列名, value_name:对应值的列名 
# 长表转换成宽表
pd.pivot(df2, index = ['id', 'name'], columns = '科目', values = '分数')
```

```python
# 数据分箱 pd.cut(x, bins, labels)
df = pd.read_csv('data.csv')
df
```

```python
df1 = df[['name', 'score', 'height']]
df1
```

```python
pd.cut(df1['score'], bins = 2) # bins = n:分割为n个区间,起始值为最小值,结束值是最大值
pd.cut(df1['score'], bins = 2).value_counts()
pd.cut(df1['score'], bins = [0, 60, 80, 100]) # bins = []:按列表中的数值分割
pd.cut(df1['score'], bins = [0, 60, 80, 100], labels = ['不及格', '及格', '优']) # 分割后加标签
pd.qcut(df1['height'],3).value_counts() # 等频划分
# 字符串直接转为category后统计,数值先分箱再做统计
```

## 7.时间数据处理


```python
d = pd.Timestamp('2015-02-28 10:22')
print(d)
print(type(d))
print('年:', d.year)
print(d.hour, d.minute, d.second)
# 季度
print(d.quarter)
print(d.is_month_end)
# methods
print(d.day_name()) # 星期几
print(d.to_period("D")) # 转换为天/季度等
print(d.to_period("Q"))
print(d.to_period("W"))
```

```python
# str to Timestap
a = pd.to_datetime('2015-05-02 10:22')
print(a.day_name())

# DataFrame to Timestamp
df = pd.DataFrame({
    'sales':[100,200,300],
    'date':['20250601','20250602','20250603']
})
df['datetime'] = pd.to_datetime(df['date'])
df
print(df.info())
# .dt: Series to Timestamp
df['dayname'] = (df['datetime'].dt.day_name())
df
df['datetime'].dt.year
```

## 8.分组聚合

```python
# df.groupby('column')['target_column'].function
df = pd.read_csv('data.csv')
df.groupby('gender').groups # 查看分组
df.groupby('gender').get_group('male') # 查看特定分组
df1 = df.groupby('gender')[['score']].mean()
df1 = df1.reset_index()
df1.sort_values('score', ascending = False)
# 根据多个条件分组
df1 = df.groupby(['gender', 'height'])[['score']].mean()
df2 = df1.reset_index()
df2.sort_values('score', ascending = True)
```
